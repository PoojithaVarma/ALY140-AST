```python
bs_object = bs(htmlfile('grativational_wave.html'), 'html.parser')
number_of_times = 0
for para in bs_object.find_all("p"):
  para1 = para.text.lower()
  if "gravitational wave" in para1:
    number_of_times = number_of_times + 1

print("Number of paragraphs in file is: {}".format(len(bs_object.find_all("p"))))
print("Number of paragraphs containing text gravitational wave is: {}".format(number_of_times))

```
